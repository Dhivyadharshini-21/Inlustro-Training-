```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
digits=load_digits()
```

```python
df=pd.DataFrame(digits.data)
df['target']=digits.target
print(df.head())
```

```python
kmeans=KMeans(n_clusters=10,random_state=42)
kmeans.fit(digits.data)
```

```python
df['cluster']=kmeans.labels_
print(df.head())
```

```python
pca=PCA(n_components=2)
pca_result=pca.fit_transform(digits.data)
plt.scatter(pca_result[:,0],pca_result[:,1],c=df['cluster'],cmap='viridis')
plt.xlabel('PCA componenent 1')
plt.ylabel('PCA componenent 2')
plt.title('KMeans clustering on digits dataset')
plt.show()
```

```python
cluster_centers=kmeans.cluster_centers_.reshape(10,8,8)
fig,axes=plt.subplots(2,5,figsize=(8,4))
for i,ax in enumerate(axes.flat):
  ax.imshow(cluster_centers[i],cmap='gray')
  ax.set_title(f'cluster{i}')
  ax.axis('off')
plt.show()
```
