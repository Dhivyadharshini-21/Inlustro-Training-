```python
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Age",data=df)
plt.title("Number of Age",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Gender',y='Survived',data=df,palette='dark')
plt.title('number of Survival based on Age',fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df[['Age','Fare']],palette='deep')
plt.title('Boxplot for outliners')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(x='Pclass',y='Age',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='PassengerId',y='Gender',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.pie(x='Pclass',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='Gender',y='Survived',data=df)
plt.show()
```
